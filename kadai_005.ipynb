```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)
#line graph
y = [1, 2, -5, 2]
axes[0, 0].plot(y)
axes[0, 0].set_title('line graph')

# sin function
x1 = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x1)
axes[0, 1].plot(x1, y1)
axes[0, 1].set_title('sin function')

# histogram
data = np.random.randn(1000)
axes[0, 2].hist(data, bins=30, color='skyblue', edgecolor='black')
axes[0, 2].set_title('histogram')

# scatter
x3 = np.random.randn(100)
y3 = 2 * x3 + np.random.randn(100)
axes[1, 0].scatter(x3, y3)
axes[1, 0].set_title('scatter')


fig.suptitle('graph')
plt.show()
```
